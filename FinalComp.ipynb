# Healthy Plate Builder

Our BILD 62 final project aims to promote healthier, yet tasty, eating habits. The USA has many unhealthy food options and has oversaturated our daily meals with fast and overprocessed foods. Those options are not nutritional and beneficial for a balanced diet, so using our program would help guide people in a better direction regarding choosing what to eat. 

## About this project
By utilizing data from the USDA food database we would like to offer an interactive plate builder where users can build their plate with a selection of foods that are nutritious (fruits, veggies, protein, dairy, carbohydrates). Once a user has finished building their plate we hope to provide a nutrional analysis of the plate with details on balanced the meal is.

### Part 1: Importing, Cleaning, and Organizing Data

This part of our project we aim to complete a couple of things: 1.Succesfully import the data we'd like to use , 2. Clean our data, 3. Have our data ready for use in the following parts.

The data we are using comes from the FoodData Central of the USDA. Definitions for the terms and labels used in our chosen dataset is [here](Bild62/Data/Field_Descriptions.pdf)

We downloaded CSV files from the USDA Food Central Data site. There were 25 CSV files and we used 4 of them for the basis of our project.

As part of cleaning and organizing our data we combined our CSV files and kept information the following; food descriptions, food type, calories, proteins, fats, and carbohydrates for each food item.

### Part 2: Design our Interactive Aspects

This part of our project we aim to sort out how we would like for user input to work.

The interactive aspect of our project includes users creating their plate and offering an option to save their plate for later reference. We ask for users to add items to their plate one by one and the program will stop recording items when the user inputs 'stop'. We then allow for users save their plates by creating a username and password. 

In the main execution of our program we ask for a couple user responses in the form of yes/no and create/retrieve each options prompts the program to have a response.

### Part 3: Nutritional Analysis

This part of our project we aim to do the necessary work that will allow for user input to be translated into nutritional details.

Our nutritional analysis is given in the form of a visualization. Using a users plate we add the total nutrients for each nutrient and compare it to the FDAs daily nutritional values Calories: 2000 kcal , Protein: 50 g, Carbohydrates: 275 g, Fats: 78 g. We then display what percentage of a users plate in relation to the daily nutritional values from 0% to 100%.

### Part 4: Visualization

This part of our project we aim to neatly display the results of Parts 1 and 2 to our users. 

The additional visualization for our project shows what a users plate looks like based on the food groups of the foods inputed. 

### Part 5: Healthy Plate Builder in Action

This part of our project is where we'd like for our users to be able to use our interactive plate builder .

```python
# importing all the necessary libraries, collections, and our scripts.py

import sys
from scripts import (
    load_and_prepare_data,
    save_plate,
    retrieve_plate,
    get_nutritional_data,
    visualize_plate
)
import os

#loading our data and prepping for saved plates
foundationfood = load_and_prepare_data()
user_data_file = os.path.join("Data", "user_plates.json")

# the point of execution for our project 

def main():
    print("Welcome to the Healthy Plate Builder!\n")
    
    # Choose to create or retrieve a plate
    while True:
        choice = input("Would you like to create a new plate or retrieve a saved plate? (create/retrieve): ").strip().lower()
        if choice in ["create", "retrieve"]:
            break
        else:
            print("Invalid input. Please type 'create' or 'retrieve'.")

    # Create a new plate
    if choice == "create":
        plate = []
        print("\nAdd foods to your plate. Type 'stop' when you're done.\n")
        while True:
            food_item = input("Enter a food to add: ").strip()
            if food_item.lower() == "stop":
                break
            elif food_item:
                plate.append(food_item)
                print(f"Added '{food_item}' to your plate.")
            else:
                print("Invalid input. Try again.")

        if input("\nDo you want to save your plate? (yes/no): ").strip().lower() == "yes":
            username = input("Enter a username: ").strip()
            password = input("Enter a password: ").strip()
            save_plate(username, password, plate, user_data_file)

    # Retrieve a saved plate
    else:
        username = input("\nEnter your username: ").strip()
        password = input("Enter your password: ").strip()
        plate = retrieve_plate(username, password, user_data_file)

    if plate:
        nutritional_data = get_nutritional_data(plate, foundationfood)
        if nutritional_data:
            if input("\nView visualizations? (yes/no): ").strip().lower() == "yes":
                visualize_plate(plate, nutritional_data)
        else:
            print("No nutritional data found.")

if __name__ == "__main__":
    main()
```
